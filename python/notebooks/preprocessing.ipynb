```python
# 1.	Pembersihan Data
import pandas as pd
import numpy as np
df = pd.read_excel('../data/Data_Kota_Surabaya.xlsx')
print(df.head(21))
```

```python
df['Harga (Rp)'] = df['Harga (Rp)'].astype(str).str.replace(',', '')
df['Harga (Rp)'] = df['Harga (Rp)'].replace('-', np.nan)
df['Harga (Rp)'] = pd.to_numeric(df['Harga (Rp)'])
print(df.head(21))
```

```python
### SEL 3: Bersihkan Kolom Tanggal (Menjadi Indeks)

# 1. Konversi Kolom Tanggal (Sesuai format Anda: dd/ mm/ YYYY)
try:
    df['Tanggal'] = pd.to_datetime(df['Tanggal'], format='%d/ %m/ %Y')
except Exception as e:
    print(f"Error konversi tanggal: {e}")

# 2. Set Tanggal sebagai Indeks (Penting untuk time series)
df = df.set_index('Tanggal')
print(df.head(21))
```

```python
# Mengisi nilai kosong dengan nilai sebelumnya (Forward Fill)
df['Harga (Rp)'] = df['Harga (Rp)'].fillna(method='ffill')
df['Harga (Rp)'] = df['Harga (Rp)'].fillna(method='bfill')
print(df.head(15))
```

```python
#SURABAYA

import pandas as pd
import numpy as np
import os

# 1. LOAD DATA
df = pd.read_excel('../data/Data_Kota_Surabaya.xlsx') # Sesuaikan path

# 2. BERSIHKAN HARGA
df['Harga (Rp)'] = df['Harga (Rp)'].astype(str).str.replace(',', '')
df['Harga (Rp)'] = df['Harga (Rp)'].replace('-', np.nan)
df['Harga (Rp)'] = pd.to_numeric(df['Harga (Rp)'])

# 3. BERSIHKAN TANGGAL
try:
    df['Tanggal'] = pd.to_datetime(df['Tanggal'], format='%d/ %m/ %Y')
except Exception as e:
    print(f"Error konversi tanggal: {e}")

# Urutkan data berdasarkan tanggal (PENTING untuk time series!)
df = df.sort_values(by='Tanggal')

# Set Index
df = df.set_index('Tanggal')

# 4. ISI DATA KOSONG
df['Harga (Rp)'] = df['Harga (Rp)'].fillna(method='ffill')
df['Harga (Rp)'] = df['Harga (Rp)'].fillna(method='bfill')

# Opsional: Buang kolom tidak penting (misal 'No Komoditas' dll) biar bersih
df = df[['Harga (Rp)']] 

print("Preview Data Bersih:")
print(df.head())

# --- SIMPAN DATA BERSIH ---
folder_tujuan = r"E:\Skripsi\Prediksi-bawang-merah\python\data\data clean"
nama_file_output = "Data_Clean_Surabaya.xlsx"
full_path = os.path.join(folder_tujuan, nama_file_output)

os.makedirs(folder_tujuan, exist_ok=True) # Buat folder otomatis jika belum ada

try:
    # PERBAIKAN DISINI: Gunakan index=True agar tanggal ikut tersimpan
    df.to_excel(full_path, index=True) 
    print(f"✓ BERHASIL! Data bersih disimpan di:")
    print(f"   {full_path}")
except Exception as e:
    print(f"X Gagal menyimpan. Error: {e}")
```
