```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# ==========================================
# KONFIGURASI
# ==========================================
path_file_input = r"E:\Skripsi\Prediksi-bawang-merah\python\data\data clean\Data_Clean_Surabaya.xlsx"
WINDOW_SIZE = 7

# Fungsi Sliding Window
def create_sliding_window(dataset, window_size):
    X, Y = [], []
    for i in range(len(dataset) - window_size):
        X.append(dataset[i:(i + window_size), 0])
        Y.append(dataset[i + window_size, 0])
    return np.array(X), np.array(Y)

print("--- MEMULAI PROSES GENERATE BUKTI BAB 6 ---")

try:
    # 1. LOAD DATA
    df = pd.read_excel(path_file_input)
    
    # Deteksi kolom harga
    col_name = 'Harga (Rp)' if 'Harga (Rp)' in df.columns else df.columns[-1]
    data_harga_asli = df[col_name].values.reshape(-1, 1)

    # 2. PROSES NORMALISASI
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data_harga_asli)

    # 3. PROSES DENORMALISASI (UNTUK BUKTI)
    denormalized_data = scaler.inverse_transform(scaled_data)

    # 4. PROSES SLIDING WINDOW (UNTUK BUKTI DIMENSI)
    X_scaled, Y_scaled = create_sliding_window(scaled_data, WINDOW_SIZE)
    X_scaled_3D = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

    # 5. PROSES SLIDING WINDOW (UNTUK BUKTI VISUAL MANUSIA)
    X_asli, Y_asli = create_sliding_window(data_harga_asli, WINDOW_SIZE)

    # ==========================================
    # OUTPUT HASIL (SCREENSHOT BAGIAN INI)
    # ==========================================

    # --- BAGIAN A: NORMALISASI ---
    print("\n" + "="*60)
    print("   BUKTI 1: HASIL NORMALISASI (Sub-bab 6.1.3)")
    print("="*60)
    print("Keterangan: Mengubah range harga asli ke range 0-1")
    print("-" * 60)
    
    # Buat tabel perbandingan
    df_norm = pd.DataFrame({
        'Tanggal': df['Tanggal'].head(10), # Asumsi ada kolom tanggal
        'Harga Asli (Rp)': data_harga_asli[:10].flatten(),
        'Hasil Normalisasi (0-1)': scaled_data[:10].flatten()
    })
    print(df_norm.to_string(index=False))

    # --- BAGIAN B: DENORMALISASI ---
    print("\n" + "="*60)
    print("   BUKTI 2: HASIL DENORMALISASI / INVERSE (Validasi)")
    print("="*60)
    print("Keterangan: Membuktikan data 0-1 bisa kembali ke Rupiah dengan tepat")
    print("-" * 60)
    
    # Buat tabel validasi
    df_denorm = pd.DataFrame({
        'Input Normalisasi': scaled_data[:10].flatten(),
        'Hasil Denormalisasi (Rp)': denormalized_data[:10].flatten(),
        'Cek Sama Dgn Asli?': np.isclose(data_harga_asli[:10], denormalized_data[:10]).flatten()
    })
    # Format tampilan float biar rapi
    pd.set_option('display.float_format', lambda x: '%.6f' % x)
    print(df_denorm.to_string(index=False))


    # --- BAGIAN C: STRUKTUR DATA LSTM ---
    print("\n" + "="*60)
    print("   BUKTI 3: TRANSFORMASI DIMENSI (Sub-bab 6.1.4)")
    print("="*60)
    print(f"Total Data Baris : {len(df)}")
    print(f"Shape Input (X)  : {X_scaled_3D.shape}")
    print("Artinya: (Jumlah Sampel, 7 Hari Mundur, 1 Fitur)")
    

    # --- BAGIAN D: VISUALISASI SLIDING WINDOW ---
    print("\n" + "="*60)
    print("   BUKTI 4: LOGIKA SLIDING WINDOW (Sub-bab 6.1.4)")
    print("="*60)
    print("Menampilkan logika pergeseran data (H-7 s/d H-1 untuk target H+1)")
    print("-" * 60)
    
    jumlah_view = 10
    cols = [f'H-{i}' for i in range(WINDOW_SIZE, 0, -1)] + ['Target']
    data_visual = np.hstack([X_asli[:jumlah_view], Y_asli[:jumlah_view].reshape(-1, 1)])
    df_vis = pd.DataFrame(data_visual, columns=cols)

    # Format ribuan (Rp)
    print(df_vis.applymap(lambda x: '{:,.0f}'.format(x)).to_string(index=False))

except Exception as e:
    print(f"Error: {e}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# ==========================================
# KONFIGURASI
# ==========================================
path_file_input = r"E:\Skripsi\Prediksi-bawang-merah\python\data\data clean\Data_Clean_Surabaya.xlsx"
WINDOW_SIZE = 7

# Fungsi Sliding Window
def create_sliding_window(dataset, window_size):
    X, Y = [], []
    for i in range(len(dataset) - window_size):
        X.append(dataset[i:(i + window_size), 0])
        Y.append(dataset[i + window_size, 0])
    return np.array(X), np.array(Y)

print("--- MEMULAI PROSES GENERATE VISUALISASI HARGA ASLI ---")

try:
    # 1. LOAD DATA
    df = pd.read_excel(path_file_input)
    col_name = 'Harga (Rp)' if 'Harga (Rp)' in df.columns else df.columns[-1]
    data_harga_asli = df[col_name].values.reshape(-1, 1)

    # 2. PROSES SLIDING WINDOW PADA DATA ASLI (RUPIAH)
    # Kita langsung pakai data asli, tidak perlu lewat scaler untuk visualisasi ini
    X_asli, Y_asli = create_sliding_window(data_harga_asli, WINDOW_SIZE)

    # ==========================================
    # VISUALISASI HEATMAP (DATA RUPIAH)
    # ==========================================
    print("\n" + "="*60)
    print("   GENERATE VISUALISASI HEATMAP (HARGA ASLI/RUPIAH)...")
    print("="*60)

    # 1. Konfigurasi Data Plot
    # Ambil 10 sampel pertama agar kotak terlihat besar dan jelas
    n_samples_plot = 10 
    X_plot = X_asli[:n_samples_plot]
    
    # 2. Setup Plot
    fig, ax = plt.subplots(figsize=(14, 9)) # Ukuran diperbesar sedikit

    # 3. Buat Heatmap
    # Gunakan cmap 'YlGnBu' (Kuning-Hijau-Biru) agar terlihat bersih untuk uang
    heatmap = ax.imshow(X_plot, aspect='auto', cmap='YlGnBu') 

    # Tambahkan Colorbar (Legenda Warna) dengan format Rupiah
    cbar = fig.colorbar(heatmap)
    cbar.set_label('Harga Bawang Merah (Rp)', rotation=270, labelpad=25, fontsize=12, fontweight='bold')

    # 4. Konfigurasi Sumbu (Labels)
    
    # Sumbu X: Lag Days (t-7 s/d t-1)
    ax.set_xticks(np.arange(WINDOW_SIZE))
    ax.set_xticklabels([f't-{WINDOW_SIZE-i}' for i in range(WINDOW_SIZE)], fontsize=11, fontweight='bold')
    ax.set_xlabel("Fitur Waktu (Posisi Hari dalam Window)", fontsize=12, labelpad=15)
    ax.xaxis.tick_top() # Label ditaruh di atas
    ax.xaxis.set_label_position('top')

    # Sumbu Y: Sampel ke-berapa
    ax.set_yticks(np.arange(n_samples_plot))
    ax.set_yticklabels([f'Sample {i}' for i in range(n_samples_plot)], fontsize=11)
    ax.set_ylabel("Urutan Sampel Data (Waktu Berjalan Maju ↓)", fontsize=12, labelpad=15)

    # 5. MENAMPILKAN ANGKA RUPIAH DI DALAM KOTAK
    # Hitung nilai tengah untuk menentukan warna teks (agar kontras)
    val_min = X_plot.min()
    val_max = X_plot.max()
    mid_point = (val_max + val_min) / 2

    for i in range(n_samples_plot):
        for j in range(WINDOW_SIZE):
            val = X_plot[i, j]
            
            # Logika warna teks: Jika background gelap (nilai tinggi), teks putih.
            # Jika background terang (nilai rendah), teks hitam.
            text_color = "white" if val > mid_point else "black"
            
            # Format angkanya dengan pemisah ribuan (contoh: 25,000)
            ax.text(j, i, f"{int(val):,}", ha="center", va="center", 
                    color=text_color, fontsize=11, fontweight='bold')

    # 6. Judul
    ax.set_title(f"Visualisasi Transformasi Data: Sliding Window (Window Size={WINDOW_SIZE})\n"
                 f"Menunjukkan Pergeseran Data Harga Asli (Rupiah)", 
                 fontsize=15, fontweight='bold', pad=30)

    plt.tight_layout()
    plt.show()
    print("Visualisasi Heatmap Rupiah berhasil ditampilkan.")

except Exception as e:
    print(f"Error: {e}")
```

```python
import pandas as pd
import numpy as np
import os

print("--- MULAI PROSES PEMBERSIHAN DATA BANYUWANGI 2025 ---")

file_path = r"E:\Skripsi\Prediksi-bawang-merah\python\data\Data_Kab_Banyuwangi_2025.xlsx"

if os.path.exists(file_path):
    # Membaca file excel ke dalam variabel df
    df = pd.read_excel(file_path)
    print("✓ File ditemukan dan berhasil dimuat!")
    
    df.columns = df.columns.str.strip()
    print(f"Nama Kolom terdeteksi: {df.columns.tolist()}")

    # 2. BERSIHKAN HARGA
    # Pastikan kolom Harga (Rp) ada sebelum diproses
    if 'Harga (Rp)' in df.columns:
        # Ubah ke string dulu, hapus koma, ganti '-' jadi NaN
        df['Harga (Rp)'] = df['Harga (Rp)'].astype(str).str.replace(',', '', regex=False)
        df['Harga (Rp)'] = df['Harga (Rp)'].replace(['-', 'nan', 'None'], np.nan)
        
        # Konversi ke angka (coerce akan mengubah yang error jadi NaN)
        df['Harga (Rp)'] = pd.to_numeric(df['Harga (Rp)'], errors='coerce')
    else:
        print("⚠ Kolom 'Harga (Rp)' tidak ditemukan! Cek nama kolom di Excel.")

    # 3. BERSIHKAN TANGGAL
    if 'Tanggal' in df.columns:
        try:
            # Menggunakan dayfirst=True agar python paham format Indonesia (DD/MM/YYYY)
            df['Tanggal'] = pd.to_datetime(df['Tanggal'], dayfirst=True, errors='coerce')
        except Exception as e:
            print(f"Error konversi tanggal: {e}")

        # Hapus baris yang tanggalnya NaT (Not a Time/Gagal konversi)
        df = df.dropna(subset=['Tanggal'])

        # Urutkan data berdasarkan tanggal (PENTING untuk time series!)
        df = df.sort_values(by='Tanggal')

        # Set Index
        df = df.set_index('Tanggal')
    else:
        print("⚠ Kolom 'Tanggal' tidak ditemukan! Pastikan nama kolom di Excel benar (Case Sensitive).")

    # 4. ISI DATA KOSONG (Missing Values)
    # Menggunakan ffill() dan bfill() yang lebih baru
    df['Harga (Rp)'] = df['Harga (Rp)'].ffill()
    df['Harga (Rp)'] = df['Harga (Rp)'].bfill()

    # Ambil hanya kolom harga
    df = df[['Harga (Rp)']] 

    print("\nPreview Data Bersih:")
    print(df.head())
    print(f"Jumlah baris data: {len(df)}")

    # 5. SIMPAN DATA BERSIH
    folder_tujuan = r"E:\Skripsi\Prediksi-bawang-merah\python\data\data clean"
    nama_file_output = "Data_Clean_Banyuwangi_2025.xlsx"
    full_path = os.path.join(folder_tujuan, nama_file_output)

    os.makedirs(folder_tujuan, exist_ok=True) 

    try:
        # Gunakan index=True agar kolom Tanggal (yang jadi index) ikut tersimpan
        df.to_excel(full_path, index=True) 
        print(f"\n✓ SUKSES! Data bersih disimpan di:")
        print(f"   {full_path}")
    except Exception as e:
        print(f"\nX Gagal menyimpan. Error: {e}")
        # Jika gagal save karena permission, coba simpan dengan nama beda
        alt_path = full_path.replace(".xlsx", "_v2.xlsx")
        df.to_excel(alt_path, index=True)
        print(f"   -> Disimpan dengan nama alternatif: {alt_path}")

else:
    print(f"❌ FILE TIDAK DITEMUKAN DI: {file_path}")
    print("Cek kembali nama file dan folder Anda.")
```
